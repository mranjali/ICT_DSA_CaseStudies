# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
#find the number of male and female buyers participated in sales and plotted on a bar graph
maleb=cars_data[cars_data['Buyer Gender']=='Male'].index
femaleb=cars_data[cars_data['Buyer Gender']=='Female'].index
malenum=len(maleb)
femalenum=len(femaleb)
print("Buyer Male:",malenum,"\nBuyer Female:",femalenum)
```

```python
buygendf={'Buyer Gender':['Male','Female'],'Buyer Num':[malenum,femalenum]}
BG=pd.DataFrame(buygendf)

c=['magenta','orange']
BG.plot(x='Buyer Gender',y='Buyer Num',kind='bar',color=c,rot=0,legend=None)
plt.grid(linestyle='dotted',color='green')
plt.suptitle(' Male vs Female buyers participated in the sales',fontsize=16)
plt.xlabel('Gender participated in sales')
plt.ylabel('Number of Buyer by gender')
plt.show()
```

# 4. Find the top 5 cars based on their sales price.

```python
#Sort the dataset based on sale price and display top 5 cars
data=cars_data.sort_values('Sale Price',ascending=False).head()
d1=pd.DataFrame(data,columns=['Make','Model','Sale Price'])
d1
```
