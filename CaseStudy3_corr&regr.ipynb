```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv('Add_sales_data_2018.csv')
```

# 1. First of all, check whether there is any relationship between the amount spent for advertising through five different companies and profit generated.

```python
corrmatrix=data.corr()
plt.subplots(figsize=(10,4))
sns.heatmap(corrmatrix,annot=True,linewidth=0.2,vmin=.9,vmax=1,cmap='YlGnBu')
```

From the correlation matrix heatmap , we can see that there is strong positive correlation between Advertisement company-2 and profit with 1.
We can see that profit having the highest correlation so our target variable(Dependent varaiable) should be profit and advertising companies are the independent variables

# 2. Build a regression model to predict the profit from the given data and comment your findings on it.

```python
#adv=['Add1(in Thousands)', 'Add2(in Thousands)','Add3(in Thousands)', 'Add4(in Thousands)', 'Add5(in Thousands)']
y=data['Profit(in Thousands)']
x=data.drop(['Date','Profit(in Thousands)'],axis=1)

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
from sklearn import linear_model
lr=linear_model.LinearRegression()
```

```python
model=lr.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
```

```python
print(y_test)
```

```python
print(predictions)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print('MSE is',mean_squared_error(y_test,predictions))
```

minimize this value as much as possible, to make the model perfect.
